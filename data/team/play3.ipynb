```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
import numpy as np
```

```python
deflated_df = deflated_df.dropna().sort_index()
streak_df = streak_df.loc[deflated_df.index].sort_index()
df = df.loc[deflated_df.index].sort_index()
```

```python
# All features being used

# Old features
WL = ['WL_for_prev_0', 
      'WL_for_prev_0_opp']
AST = ['AST_for_prev_0', 'AST_against_prev_0', 
       'AST_for_prev_0_opp', 'AST_against_prev_0_opp']
REB = ['REB_for_prev_0', 'REB_against_prev_0', 
       'REB_for_prev_0_opp', 'REB_against_prev_0_opp']
FG_PCT = ['FG_PCT_for_prev_0', 'FG_PCT_against_prev_0', 
          'FG_PCT_for_prev_0_opp', 'FG_PCT_against_prev_0_opp']
FG3_PCT = ['FG3_PCT_for_prev_0', 'FG3_PCT_against_prev_0', 
           'FG3_PCT_for_prev_0_opp', 'FG3_PCT_against_prev_0_opp']
PTS = ['PTS_for_prev_0', 'PTS_against_prev_0', 
       'PTS_for_prev_0_opp', 'PTS_against_prev_0_opp']
old_features = WL + AST + REB + FG_PCT + FG3_PCT + PTS


# New features
PLUS_MINUS = ['PLUS_MINUS_for_prev_0', 
              'PLUS_MINUS_for_prev_0_opp']
BLK = ['BLK_for_prev_0', 'BLK_against_prev_0', 
       'BLK_for_prev_0_opp', 'BLK_against_prev_0_opp']
STL = ['STL_for_prev_0', 'STL_against_prev_0', 
       'STL_for_prev_0_opp', 'STL_against_prev_0_opp']
TOV = ['TOV_for_prev_0', 'TOV_against_prev_0', 
       'TOV_for_prev_0_opp', 'TOV_against_prev_0_opp']
PF = ['PF_for_prev_0', 'PF_against_prev_0', 
      'PF_for_prev_0_opp', 'PF_against_prev_0_opp']
MIN = ['MIN_for_prev_0', 
       'MIN_for_prev_0_opp']
OREB = ['OREB_for_prev_0', 'OREB_against_prev_0', 
        'OREB_for_prev_0_opp', 'OREB_against_prev_0_opp']
DREB = ['DREB_for_prev_0', 'DREB_against_prev_0', 
        'DREB_for_prev_0_opp', 'DREB_against_prev_0_opp']
new_features1 = PLUS_MINUS + BLK + STL + TOV + PF + MIN + OREB + DREB


FGM = ['FGM_for_prev_0', 'FGM_against_prev_0', 
       'FGM_for_prev_0_opp', 'FGM_against_prev_0_opp']
FGA = ['FGA_for_prev_0', 'FGA_against_prev_0', 
       'FGA_for_prev_0_opp', 'FGA_against_prev_0_opp']
FG3M = ['FG3M_for_prev_0', 'FG3M_against_prev_0',
        'FG3M_for_prev_0_opp', 'FG3M_against_prev_0_opp']
FG3A = ['FG3A_for_prev_0', 'FG3A_against_prev_0', 
        'FG3A_for_prev_0_opp', 'FG3A_against_prev_0_opp']
FTM = ['FTM_for_prev_0', 'FTM_against_prev_0', 
       'FTM_for_prev_0_opp', 'FTM_against_prev_0_opp']
FTA = ['FTA_for_prev_0', 'FTA_against_prev_0', 
       'FTA_for_prev_0_opp', 'FTA_against_prev_0_opp']
FT_PCT = ['FT_PCT_for_prev_0', 'FT_PCT_against_prev_0', 
          'FT_PCT_for_prev_0_opp', 'FT_PCT_against_prev_0_opp']
new_features2 = FGM + FGA + FG3M + FG3A + FTM + FTA + FT_PCT

IS_HOME = ['IS_HOME_for_prev_0', 'IS_HOME_for_prev_0_opp',
                    'IS_HOME_against_prev_0', 'IS_HOME_against_prev_0_opp',
                    'IS_HOME_for', 'IS_HOME_for_opp',
                    'IS_HOME_against', 'IS_HOME_against_opp']
```

```python
main_df = pd.concat([streak_df, df], axis=1)
```

```python
features = old_features + new_features1 + new_features2
Xs = main_df[features].values
ys = main_df['WL_for'].values
```

```python
num_states = 100
print("Running {} times".format(num_states))
test_accuracies = np.zeros(num_states)
for state in range(0, num_states):

    # Splitting data into training and testing data
    X_train, X_test, y_train, y_true = train_test_split(Xs, ys, test_size=0.2, random_state=state)

    # Model training
    lr_model = LogisticRegression(max_iter=10000, random_state=state)
    lr_model.fit(X_train, y_train)

    # Get training accuracy
    print("Accuracy run {}: ".format(state+1))
    print("Train {:.10f}".format(accuracy_score(lr_model.predict(X_train), y_train)))
    
    # Get test accuracy
    y_logits = lr_model.predict(X_test)
    y_pred = y_logits.round()
    print("Test {:.10f}".format(accuracy_score(y_pred, y_true)))
    test_accuracies[state] = accuracy_score(y_pred, y_true)
print("Test accuracies: ", test_accuracies)
print("Mean: ", test_accuracies.mean())
```

```python
seasons_arr = np.arange(2015, 2024, 1)
res = []

for season in seasons_arr:

    # Season to train for
    season += 20000
    recency_bias = 10
    recency_weight = 1

    print(' * Training model data before {} season '.format(season - 20000))
    # Splitting data into training and testing data
    print(' * Splitting data ... ')
    train_cond = (main_df['SEASON_ID'] < season) & (main_df['SEASON_ID'] >= season-10)
    X_train = main_df[train_cond][features]
    y_train = main_df[train_cond]['WL_for']
    
    """seasons_to_drop = [22020, 22021]
    X_train = X_train[~main_df['SEASON_ID'].isin(seasons_to_drop)]
    y_train = y_train.loc[~main_df['SEASON_ID'].isin(seasons_to_drop)]"""
    
    
    test_cond = (main_df['SEASON_ID'] == season) 
    X_test = main_df[test_cond][features]
    y_true = main_df[test_cond]['WL_for']

    print(' * Identifying recent data (past {} seasons)'.format(recency_bias))
    print(' * Counting recent data {} times more in training data'.format(recency_weight))



    state = 100

    # Model training
    lr_model = LogisticRegression(max_iter=10000, random_state=state)
    lr_model.fit(X_train, y_train)
    

    # Get training accuracy
    print("Train {:.10f}".format(accuracy_score(lr_model.predict(X_train), y_train)))
        
    # Get test accuracy
    y_logits = lr_model.predict(X_test)
    y_pred = y_logits.round()
    print("Test {:.10f}".format(accuracy_score(y_pred, y_true)))
    res.append(accuracy_score(y_pred, y_true))

print(seasons_arr)
print(np.array(res).round(2))
```
